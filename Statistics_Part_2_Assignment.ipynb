**1.What is hypothesis testing in statistics?**

ans-Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It involves formulating a hypothesis, collecting data, and using statistical techniques to determine whether there is enough evidence to support or reject the hypothesis.

**Key Steps in Hypothesis Testing:**

- Formulate Hypotheses- Null Hypothesis (H‚ÇÄ): Assumes there is no effect or difference.
- Alternative Hypothesis (H‚ÇÅ): Suggests there is an effect or difference.

- Choose a Significance Level (Œ±)- Common values are 0.05 or 0.01, which indicate the probability of rejecting the null hypothesis when it is actually true.

- Select a Test Statistic- Examples: Z-test, T-test, Chi-square test, ANOVA, etc.

- Compute the Test Statistic & P-value- The p-value helps determine the strength of the evidence against H‚ÇÄ.

- Make a Decision- If the p-value is less than Œ±, reject H‚ÇÄ (suggests evidence for H‚ÇÅ).
- If the p-value is greater than Œ±, fail to reject H‚ÇÄ (not enough evidence for H‚ÇÅ).







**2.What is the null hypothesis, and how does it differ from the alternative hypothesis?**

ans-The null hypothesis (H‚ÇÄ) and alternative hypothesis (H‚ÇÅ) are the two core components of hypothesis testing in statistics.

**Null Hypothesis (H‚ÇÄ):**
This hypothesis states that there is no effect, difference, or relationship between variables. It represents the default assumption, meaning that any observed results are due to chance.
- Example: A new teaching method does not improve student test scores.

**Alternative Hypothesis (H‚ÇÅ):**
This hypothesis contradicts the null hypothesis and suggests that there is an effect, difference, or relationship between variables.
- Example: A new teaching method significantly improves student test scores.




**3.What is the significance level in hypothesis testing, and why is it important?**

ans-The significance level (Œ±) in hypothesis testing is the probability of rejecting the null hypothesis (H‚ÇÄ) when it is actually true. It acts as a threshold to determine whether the observed result is statistically significant.

**Why Is It Important?**

- Controls Type I Error: It helps limit the chance of mistakenly rejecting H‚ÇÄ when it is actually correct.
- Establishes Confidence: A lower Œ± (e.g., 0.01) makes the test more stringent, requiring stronger evidence to reject H‚ÇÄ.
- Influences Decision-Making: The significance level helps researchers determine the reliability of their findings.




**4.What does a P-value represent in hypothesis testing?**

- A small p-value (typically ‚â§ Œ±) suggests strong evidence against H‚ÇÄ, leading to its rejection.
- A large p-value means there is not enough evidence to reject H‚ÇÄ.

**6.What are Type 1 and Type 2 errors in hypothesis testing?**

ans-Type 1 and Type 2 errors are fundamental concepts in hypothesis testing that describe mistakes in decision-making:

* Type 1 Error (False Positive): This occurs when you reject the null hypothesis when it is actually true. In simpler terms, it's like convicting an innocent person in court. The probability of making this error is denoted by Œ± (alpha), which represents the significance level.

* Type 2 Error (False Negative): This happens when you fail to reject the null hypothesis when it is actually false. Think of it as letting a guilty person go free. The probability of making this error is denoted by Œ≤ (beta).

Balancing these errors is key in statistical analysis. Reducing the likelihood of a Type 1 error usually increases the risk of a Type 2 error, and vice versa. The choice depends on the context‚Äîwhether avoiding false positives or false negatives is more critical.

**14.What is a confidence interval, and how is it used to interpret statistical results?**

ans-A confidence interval (CI) is a range of values used to estimate an unknown population parameter, such as a mean or proportion, based on sample data. It provides a margin of error around an estimate and indicates the level of certainty about the result.

**Key Components of a Confidence Interval**
* Point Estimate: The sample statistic (e.g., sample mean) used as the best guess for the population parameter.

* Margin of Error: Accounts for sampling variability and is influenced by the confidence level and sample size.

* Confidence Level: The probability that the true population parameter lies within the interval. Common levels are 90%, 95%, or 99%.

**15.What is the margin of error, and how does it affect the confidence interval?**

ans-The margin of error (MOE) represents the range within which the true population parameter is expected to lie, given a sample. It's crucial in statistical estimates, particularly confidence intervals.

**How It Works:**
The margin of error is calculated as:
[ MOE = Z \times \frac{\sigma}{\sqrt{n}} ]
Where:
- Z is the Z-score (based on confidence level, e.g., 1.96 for 95% confidence).
- œÉ is the population standard deviation.

**Impact on Confidence Interval:**
A confidence interval is given by:
[ \text{Sample Estimate} \pm \text{Margin of Error} ]
- A larger MOE means a wider confidence interval, indicating more uncertainty.
- A smaller MOE results in a narrower confidence interval, meaning a more precise estimate.

Factors influencing MOE include:
- Sample Size (n): Larger samples reduce the margin of error.
- Confidence Level: Higher confidence levels increase MOE.
- Population Variability (œÉ): More variability leads to a larger MOE.

**20.What is an ANOVA test, and what are its assumptions?**

ans-ANOVA (Analysis of Variance) Test
ANOVA is a statistical method used to compare the means of multiple groups to determine if they are significantly different from one another. Instead of comparing groups pairwise (like with t-tests), ANOVA evaluates variability between and within groups simultaneously.

**Types of ANOVA:**
- One-Way ANOVA: Used when comparing means of one independent variable across multiple groups.
- Two-Way ANOVA: Used when analyzing two independent variables and their interactions.

**Assumptions of ANOVA:**
To ensure reliable results, ANOVA relies on the following assumptions:
- Independence: Observations must be independent of one another.
- Normality: Each group‚Äôs data should follow a normal distribution.
- Homogeneity of Variance (Homoscedasticity): The variances of the different groups should be approximately equal.
- Random Sampling: Data should be collected using random sampling methods.
- Continuous Data: The dependent variable should be measured on a continuous scale.




**21.What are the different types of ANOVA tests?**

ans-ANOVA (Analysis of Variance) comes in different types depending on the number of independent variables and the complexity of the study. Here are the main types:
1. One-Way ANOVA
- Used to compare the means of three or more groups based on one independent variable.
- Example: Comparing the test scores of students across three different schools.

2. Two-Way ANOVA
- Used when there are two independent variables and their possible interaction effects.
- Example: Studying how diet (A) and exercise level (B) affect body weight.

3. Repeated Measures ANOVA
- Used when the same subjects are measured multiple times under different conditions.
- Example: Tracking cognitive performance before, during, and after a new training program.

4. MANOVA (Multivariate ANOVA)
- Extends ANOVA to multiple dependent variables instead of just one.
- Example: Analyzing sales revenue and customer satisfaction across multiple stores.

5. ANCOVA (Analysis of Covariance)
- A modification of ANOVA that controls for a covariate, a variable that might influence the outcome.
- Example: Studying how marketing strategies impact sales, while adjusting for store location.

**2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define possible values
x = np.arange(1, 11)  # Values from 1 to 10
pmf_values = np.ones_like(x) / len(x)  # Equal probability for each value

# Plot the PMF
plt.bar(x, pmf_values, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('PMF of Discrete Uniform Distribution')
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution**

```python
import numpy as np
import matplotlib.pyplot as plt

def bernoulli_pdf(p):
    """Calculate and visualize the probability distribution function (PDF) of a Bernoulli distribution."""
    x = np.array([0, 1])  # Possible outcomes (0 = failure, 1 = success)
    pdf_values = np.array([1 - p, p])  # Probability mass function (PMF)

    # Plot the Bernoulli distribution
    plt.bar(x, pdf_values, color='skyblue', edgecolor='black', alpha=0.7)
    plt.xticks(x, labels=['Failure (0)', 'Success (1)'])
    plt.xlabel('Outcome')
    plt.ylabel('Probability')
    plt.title(f'Bernoulli Distribution (p={p})')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Example usage
bernoulli_pdf(0.7)  # Assuming probability of success p = 0.7
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of values
x = np.arange(1, 11)  # Values from 1 to 10
cdf_values = np.cumsum(np.ones_like(x) / len(x))  # CDF calculation

# Plot the CDF
plt.step(x, cdf_values, where='mid', color='skyblue', linewidth=2, label='CDF')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF) of Discrete Uniform Distribution')
plt.xticks(x)
plt.yticks(np.linspace(0, 1, len(x)))
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

**12.Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal():
    """Generate and plot the standard normal distribution (mean=0, std=1)."""
    x = np.linspace(-4, 4, 1000)  # Range of values
    pdf_values = norm.pdf(x, 0, 1)  # Probability density function

    # Plot the distribution
    plt.plot(x, pdf_values, color='skyblue', linewidth=2, label='Standard Normal Distribution')
    plt.xlabel('Value')
    plt.ylabel('Probability Density')
    plt.title('Standard Normal Distribution (Mean=0, Std Dev=1)')
    plt.grid(linestyle='--', alpha=0.7)
    plt.legend()
    plt.show()

# Call the function
plot_standard_normal()
```

**14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(value, data):
    """Calculate Z-score for a given value based on dataset statistics."""
    mean = np.mean(data)
    std_dev = np.std(data)
    return (value - mean) / std_dev

# Sample dataset (normally distributed)
data = np.random.normal(50, 10, 1000)  # Mean=50, Std Dev=10
value = 60  # Data point to analyze

# Compute Z-score
z_score = calculate_z_score(value, data)
print(f"Z-score of {value}: {z_score:.2f}")

# Plot standard normal distribution
x = np.linspace(-4, 4, 1000)
pdf_values = norm.pdf(x, 0, 1)

plt.plot(x, pdf_values, color='skyblue', linewidth=2, label='Standard Normal Distribution')
plt.axvline(z_score, color='red', linestyle='--', label=f'Z-score of {value}')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.title('Standard Normal Distribution with Highlighted Z-score')
plt.legend()
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

**5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(alpha=0.05, beta=0.2, mean_null=50, mean_alt=53, std_dev=10):
    """
    Visualizes Type 1 and Type 2 errors in hypothesis testing.

    Parameters:
    alpha (float)   : Significance level (Type 1 error probability)
    beta (float)    : Probability of Type 2 error
    mean_null (int) : Mean under null hypothesis (H0)
    mean_alt (int)  : Mean under alternative hypothesis (H1)
    std_dev (int)   : Standard deviation of both distributions
    """

    # Define x-axis range
    x = np.linspace(30, 70, 1000)

    # Compute normal distributions for H0 and H1
    pdf_null = norm.pdf(x, mean_null, std_dev)
    pdf_alt = norm.pdf(x, mean_alt, std_dev)

    # Compute critical value for Type 1 error (right-tailed)
    critical_value = norm.ppf(1 - alpha, mean_null, std_dev)

    # Plot both distributions
    plt.plot(x, pdf_null, label='Null Distribution (H0)', color='blue')
    plt.plot(x, pdf_alt, label='Alternative Distribution (H1)', color='red')

    # Shade Type 1 Error (Alpha)
    plt.fill_between(x, pdf_null, where=(x >= critical_value), color='blue', alpha=0.3, label='Type 1 Error (Œ±)')

    # Shade Type 2 Error (Beta)
    plt.fill_between(x, pdf_alt, where=(x < critical_value), color='red', alpha=0.3, label='Type 2 Error (Œ≤)')

    # Labels and legend
    plt.axvline(critical_value, color='black', linestyle='--', label=f'Critical Value (Œ±={alpha})')
    plt.xlabel('Test Statistic')
    plt.ylabel('Probability Density')
    plt.title('Type 1 & Type 2 Errors in Hypothesis Testing')
    plt.legend()
    plt.grid(linestyle='--', alpha=0.6)
    plt.show()

# Run the visualization function
visualize_errors()
```

**9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance**

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    - data: List or NumPy array of sample values
    - confidence: Confidence level (default is 95%)

    Returns:
    - Tuple containing (lower bound, upper bound)
    """
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation
    n = len(data)

    # Use t-distribution for small samples, Z-distribution for large samples
    if n < 30:
        critical_value = stats.t.ppf((1 + confidence) / 2, df=n-1)
    else:
        critical_value = stats.norm.ppf((1 + confidence) / 2)

    margin_of_error = critical_value * (sample_std / np.sqrt(n))

    return sample_mean - margin_of_error, sample_mean + margin_of_error

# Example usage
sample_data = np.random.normal(loc=50, scale=10, size=25)  # Simulated sample
ci_lower, ci_upper = confidence_interval(sample_data)
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

**11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process**

```python
def bayes_theorem(prior, likelihood, evidence):
    """
    Compute posterior probability using Bayes' Theorem.

    Parameters:
    - prior: P(H) - Initial belief (probability of disease before test)
    - likelihood: P(E|H) - Probability of a positive test given disease
    - evidence: P(E) - Overall probability of a positive test

    Returns:
    - Posterior probability P(H|E)
    """
    posterior = (likelihood * prior) / evidence
    return posterior

# Given data:
prior = 0.01  # 1% of people have the disease (P(H))
likelihood = 0.95  # 95% test sensitivity (P(E|H))
false_positive_rate = 0.05  # 5% false positive rate (P(E|¬¨H))

# Compute P(E): Total probability of testing positive
evidence = (likelihood * prior) + (false_positive_rate * (1 - prior))

# Calculate posterior probability
posterior = bayes_theorem(prior, likelihood, evidence)
print(f"Probability of having the disease given a positive test result: {posterior:.4f}")
```

**13.Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data**

```python
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    """
    Compute expected frequencies for a Chi-square test.

    Parameters:
    - observed: 2D NumPy array or Pandas DataFrame containing observed values

    Returns:
    - Expected frequency table as a Pandas DataFrame
    """
    # Convert to NumPy array if input is a DataFrame
    observed = np.array(observed)

    # Compute row and column sums
    row_totals = observed.sum(axis=1, keepdims=True)
    col_totals = observed.sum(axis=0, keepdims=True)
    grand_total = observed.sum()

    # Calculate expected frequencies
    expected = (row_totals @ col_totals) / grand_total

    return pd.DataFrame(expected, columns=["Category 1", "Category 2"], index=["Group A", "Group B"])

# Example observed data (Contingency Table)
observed_data = np.array([[50, 30],  # Group A (Category 1 & 2)
                          [20, 40]]) # Group B (Category 1 & 2)

# Calculate expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_data)

# Display results
print("Observed Data:\n", pd.DataFrame(observed_data, columns=["Category 1", "Category 2"], index=["Group A", "Group B"]))
print("\nExpected Frequencies:\n", expected_frequencies)
```

```python
import scipy.stats as stats
import numpy as np

# Observed data (actual counts)
observed = np.array([50, 30, 20])

# Automatically compute expected frequencies
expected = np.ones_like(observed) * observed.sum() / len(observed)  # Ensures sum matches

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = stats.chisquare(observed, expected)

print(f"Chi-square Statistic: {chi2_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("The observed distribution differs significantly from the expected one.")
else:
    print("The observed distribution matches the expected distribution.")
```
